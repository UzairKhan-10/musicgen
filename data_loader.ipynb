```python
import pandas as pd
import os

df_test_1 = pd.DataFrame(columns=['file_name', 'path', 'instruments'])
wav_files, txt_files = [], []
arr = os.listdir("IRMAS-TestingData-Part3/Part3")
for file in arr:
    l = len(file)
    if file[l-3:l] == 'txt':
        txt_files.append(file)
    else:
        wav_files.append(file)
print(f'.wav files: {wav_files}')
print(f'.txt files: {txt_files}')

idx = 0
for wv, txt in zip(wav_files, txt_files):
    l = len(txt)
    instruments_file = open("IRMAS-TestingData-Part3/Part3/"+txt, 'r')
    instruments = instruments_file.readlines()
    for i in range(len(instruments)):
        instruments[i] = instruments[i][0:3]
    path = "IRMAS-TestingData-Part3/Part3/"+ wv
    row = [txt[0:l-3], path, instruments]
    df_test_1.loc[idx] = row
    idx += 1
print(f'Shape: {df_test_1.shape}')
df_test_1.head()
```

```python
final_df = pd.DataFrame(columns=['file_name', 'path', 'instruments'])
final_df = pd.concat([final_df, df_test_1])
final_df.head()
```
